```python
def get_count(series, limit=None):
    if limit != None:
        series = series.value_counts()[:limit]
    else:
        series = series.value_counts()
    
    x = series.index
    y = series/series.sum()*100
    
    return x.values,y.values

```

Conclusion-A hotel booking system is a software application that is used by hotels to manage their room inventory, rates, and reservations. The system allows hotels to manage their operations efficiently, maximize their revenue potential, and provide a streamlined booking process for guests. A hotel booking system can be integrated into the hotel's property management system or be a standalone application. It provides features such as room availability management, online booking capabilities, integration with third-party booking sites, and reporting on occupancy and revenue. The complexity and functionality of a hotel booking system can vary depending on the size and type of hotel. Overall, the use of a hotel booking system is essential for effective management of hotel operations and providing guests with a seamless booking experience.
